# **Практическая работа №10. Развертывание автоматизированной системы преобразования координатных данных**

## **Цель работы**



Развернуть автоматизированную систему преобразования координатных данных, разработанную в практической работе №7, с использованием архитектуры клиент-сервер, где бэкенд размещается на платформе render.com, а фронтенд реализуется с помощью Streamlit. Система должна принимать на вход файл Excel, обрабатывать его и возвращать отчет в формате markdown.

## **Задачи**



1. Разработать бэкенд-сервис на основе FastAPI для преобразования координатных данных
2. Создать фронтенд-приложение с использованием Streamlit для удобного взаимодействия с пользователем
3. Развернуть бэкенд-сервис на платформе render.com
4. Развернуть фронтенд-приложение на Streamlit Cloud
5. Настроить взаимодействие между фронтендом и бэкендом
6. Оформить документацию и исходный код в репозитории GitHub

## **Требования к выполнению работы**

### **Задание №1. Разработка бэкенда (FastAPI)**


1. Разработать приложение на основе FastAPI для обработки координатных данных:
   - Создать эндпоинт для загрузки Excel-файла
   - Реализовать функцию преобразования координат согласно алгоритму из практической работы №7
   - Добавить эндпоинт для возвращения результатов в формате markdown
   - Обеспечить обработку ошибок и валидацию входных данных

2. Код бэкенда должен быть структурирован следующим образом:
   - main.py - основной файл приложения FastAPI
   - requirements.txt - зависимости проекта

### **Задание №2. Разработка фронтенда (Streamlit)**



1. Создать интерфейс пользователя с помощью Streamlit:
   - Добавить возможность загрузки Excel-файла
   - Реализовать отображение параметров преобразования и их настройку
   - Создать визуализацию результатов преобразования
   - Добавить возможность скачивания отчета в формате markdown

2. Код фронтенда должен быть структурирован следующим образом:
   - app.py - основной файл приложения Streamlit
   - utils/ - директория с вспомогательными функциями (при необходимости)

### **Задание №3. Развертывание системы**



1. Развернуть бэкенд на платформе render.com:
   - Создать новый Web Service на render.com
   - Настроить переменные окружения (если необходимо)
   - Проверить работоспособность развернутого сервиса

2. Развернуть фронтенд на Streamlit Cloud:
   - Создать учетную запись на Streamlit Cloud (если отсутствует)
   - Настроить репозиторий для развертывания
   - Проверить работоспособность развернутого приложения

1. Протестировать работу системы на различных наборах данных:
   - Создать тестовый набор данных с использованием скрипта генерации

## Интегральные функции (scipy.integrate)

**scipy.integrate** —  это про численное интегрирование (методы трапеции, Симпсона, Ромберга и др.), решение обыкновенных дифференциальных уравнений (ОДУ); функции для решения интегралов.

В модуле **scipy.integrate** есть несколько функций для вычисления определённых интегралов. Наиболее простой является
функция **quad()**, которая вычисляет обычный интеграл от одной переменной и имеет следующий синтаксис:
```quad (func, a, b, …)```, где:
* func — имя подынтегральной функции
* a — нижний предел интегрирования
* b — верхний предел интегрирования

Например посчитаем интеграл: $$\int_{0}^{4}3{x}^{2}dx$$

````python
from scipy import integrate

def target_function_f(x):
    return 3.0 * x ** 2

result = integrate.quad(target_function_f, 0.0, 4.0)
print(result)
````

Функция **quad()** возвращает кортеж из двух элементов, первый — значение интеграла, второй — точность вычисления. В данном случае  получен с точностью до 13-го знака после запятой. Есть и другие методы нахождения интегралов: **trapz() (метод трапеций), sims() (метод Симпсона), romb() (метод Ромберга)** и т.д.

````python
from scipy import integrate
import numpy as np

# Определение функции и создание массивов значений
def target_function_f(x):
    return 3.0 * x ** 2

x = np.linspace(0, 4, num=500)
y = target_function_f(x)

# Вычисление результатов методами трапеций, Симпсона и Ромберга
trapz_result = integrate.trapz(y, x)
simpson_result = integrate.simpson(y, x)

x_romb = np.linspace(0, 4, num=2**8 + 1)
y_romb = target_function_f(x_romb)
romb_result = integrate.romb(y_romb, dx=(x_romb[1] - x_romb[0]))

# Точное значение интеграла
exact_value = 64

# Вычисление абсолютных ошибок
trapz_error = abs(trapz_result - exact_value)
simpson_error = abs(simpson_result - exact_value)
romb_error = abs(romb_result - exact_value)

# Вывод результатов
print(f"Метод трапеций: {trapz_result}, Абсолютная ошибка: {trapz_error}")
print(f"Метод Симпсона: {simpson_result}, Абсолютная ошибка: {simpson_error}")
print(f"Метод Ромберга: {romb_result}, Абсолютная ошибка: {romb_error}")
````

### Обыкновенное дифференциальное уравнение (ОДУ)

Также давайте посмотрим как решить простенькое **ОДУ**: $$\frac{dx}{dy}=-2y$$

Для этого есть функция **odeint()** следующего вида: ```odeint(func, y0, x, …)```, где:
* func — список функций правых частей уравнений;
* y0 — список начальных условий;
* x — массив точек, в которых необходимо найти значение искомой функции $\bar{y}(x)$;
* многоточием обозначены необязательные аргументы.

Функция **odeint()** возвращает двумерный массив типа **ndarray** значений функции $y$ во всех заданных точках $x$. Начальное значение y0 хранится в первой строке. Массив решения имеет следующую форму (shape): (len(x), len(y0)). То есть число строк равно количеству точек интегрирования, число столбцов — количеству уравнений.

Например, в случае интегрирования одного уравнения результатом работы функции **odeint()** будет **двумерный массив**, содержащий один столбец, в котором хранятся значения искомой функции, начиная с y0.

````python
from scipy import integrate
import numpy as np

def target_function(y, x):
    return -2.0 * y

xi = np.linspace(0, 1, 10)
y0 = 1.0
result = integrate.odeint(target_function, y0, xi)
print(result)
````

### Можно также решать и системы ОДУ

Пусть система ОДУ выглядят следующим образом:

$$
\begin{align*}
\frac{dy_1}{dt} &= -0.5 \cdot y_1 \\
\frac{dy_2}{dt} &= 4 - 0.3 \cdot y_2 - 0.1 \cdot y_1
\end{align*}
$$

где:
- $ y_1 $ и $ y_2 $ — функции времени $ t $
- $ \frac{dy_1}{dt} $ и $ \frac{dy_2}{dt} $ — производные $ y_1 $ и $ y_2 $ по времени соответственно.

Эти уравнения описывают изменение переменных $ y_1 $ и $ y_2 $ во времени, где первое уравнение представляет собой простое экспоненциальное затухание, а второе — более сложное уравнение, включающее взаимодействие между $ y_1 $ и $ y_2 $.

````python
import numpy as np
from scipy.integrate import solve_ivp

# Функция правых частей системы дифференциальных уравнений
def system(t, y):
    return [-0.5 * y[0], 4 - 0.3 * y[1] - 0.1 * y[0]]

# Начальные условия
y0 = [1, 0]

# Интервал интегрирования
t_span = (0, 10)

# Решение системы дифференциальных уравнений
sol = solve_ivp(system, t_span, y0, t_eval=np.linspace(0, 10, 5))

# sol.t - массив значений времени
# sol.y - массив решений системы уравнений
````

Уравнение гармонических колебаний: $\frac{d^2x}{dt^2}+\omega^2x=0$ c начальными условиями $x(0) = {x}_{0} = –5$, $\frac{dx}{dt(0)} = {v}_{0} = 0$ при $\omega=1.5$. Сведим данное уравнение второго порядка к системе двух уравнений первого порядка с помощью замены: $\frac{dx}{dt}=v, \frac{dv}{dt}=\frac{d^2x}{dt^2}$. Тогда получим: $\frac{dv}{dt}=-\omega^2x, \frac{dx}{dt}=v$ c начальными условиями $x(0) = {x}_{0}, v(0) = {v}_{0}$.

````python
from scipy import integrate
import numpy as np
from matplotlib import pyplot as plt

# частота гармонического осциллятора
w = 1.5
# вектор-функция правых частей уравнений:
# f = [f0, f1], где f0 = -w^2*x, f1 = v.
# полагается, что f зависит от (y, t), причём y - это
# список из двух чисел:
# y = [v, x]
def f(y, t):
 v = y[0]
 x = y[1]
 f0 = -w ** 2 * x
 f1 = v
 return [f0, f1]

# массив точек интегрирования
ti = np.linspace(0, 10, 50)

# начальная координата
x0 = -5.0

# начальная скорость
v0 = 0.0
# список начальных условий
y0 = [x0, v0]

# решение ОДУ
sol = integrate.odeint(f, y0, ti)

# рисунок для построения графиков функций x(t) и v(t)
fig = plt.figure()

# панель для рисования графика координаты
ax1 = fig.add_subplot(121)

# панель для рисования графика скорости
ax2 = fig.add_subplot(122)
ax1.set_xlabel(r'$t$')
ax1.set_ylabel(r'$x$')
ax1.set_title('координата')

# график x(t) - зависимость нулевого столбца вектора
# решения sol от ti
ax1.plot(ti, sol[:, 0])
ax2.set_xlabel(r'$t$')
ax2.set_ylabel(r'$v$')
ax2.set_title('скорость')

# график v(t) - зависимость первого столбца вектора
# решения sol от ti
ax2.plot(ti, sol[:, 1])

# настройка оптимального расположения панелей
plt.tight_layout()
fig.savefig("ode2.png")
````

## Интерполяция (scipy.interpolate)

Интерполяция представляет собой процесс генерации новых данных на основе набора известных значений. В библиотеке scipy.interpolate представлен широкий спектр функций для интерполяции как одномерных, так и многомерных данных. Одной из ключевых функций является interp1d(), которая позволяет интерполировать одномерные данные. Функция определяется следующим образом:
```python
fi = interp1d(x, f0, 'method', …)
```
Где interp1d() принимает массивы координат $ x $ и соответствующие значения функции $ f_0 $. В качестве метода интерполяции могут быть использованы:

- линейная интерполяция, если параметр 'method' установлен в 'linear';
- квадратичная интерполяция, если параметр 'method' установлен в 'quadratic';
- кубическая интерполяция, если параметр 'method' установлен в 'cubic';

и другие. При отсутствии явного указания параметра 'method', функция по умолчанию применяет линейную интерполяцию. Для получения дополнительной информации рекомендуется обратиться к документации библиотеки.

- Рассмотрим примеры вычислений для каждого из методов интерполяции, используя следующий набор данных:

  - Известные точки: $ (1, 1) $, $ (2, 4) $, $ (3, 9) $
  - Новая точка для интерполяции: $ x = 2.5 $

- **Интерполяция методом ближайшего соседа**:
$$ f(2.5) = f(2) = 4 $$

- **Линейная интерполяция**:
$$ f(2.5) = f(2) + \frac{(2.5 - 2)}{(3 - 2)} \cdot (f(3) - f(2)) = 4 + 0.5 \cdot (9 - 4) = 6.5 $$

- **Квадратичная интерполяция**:
Для квадратичной интерполяции нам нужно рассчитать разности $ \Delta f(x_i) $ и $ \Delta^2 f(x_i) $. Предположим, что $ \Delta f(1) = 3 $ и $ \Delta^2 f(1) = 2 $, тогда:
$$ f(2.5) = f(2) + (2.5 - 2) \Delta f(2) + \frac{(2.5 - 2)(2.5 - 3)}{2!} \Delta^2 f(1) = 4 + 0.5 \cdot 3 + \frac{0.5 \cdot (-0.5)}{2} \cdot 2 = 6.25 $$

- **Кубическая интерполяция**:
Аналогично, для кубической интерполяции нам нужны значения $ \Delta f(x_i) $, $ \Delta^2 f(x_i) $, и $ \Delta^3 f(x_i) $. Предположим, что $ \Delta f(1) = 3 $, $ \Delta^2 f(1) = 2 $, и $ \Delta^3 f(1) = 1 $, тогда:
$$ f(2.5) = f(2) + (2.5 - 2) \Delta f(2) + \frac{(2.5 - 2)(2.5 - 3)}{2!} \Delta^2 f(1) + \frac{(2.5 - 2)(2.5 - 3)(2.5 - 4)}{3!} \Delta^3 f(1) = 4 + 0.5 \cdot 3 + \frac{0.5 \cdot (-0.5)}{2} \cdot 2 + \frac{0.5 \cdot (-0.5) \cdot (-1.5)}{6} \cdot 1 = 6.375 $$

- **Сплайн-интерполяция**:
Для сплайн-интерполяции мы используем кубические полиномы для каждого интервала между точками. Предположим, что для интервала между $ (2, 4) $ и $ (3, 9) $ у нас есть сплайн:
$$ S_i(x) = 4 + 3(x - 2) + 0.5(x - 2)^2 + 0(x - 2)^3 $$
Тогда для $ x = 2.5 $:
$$ S_i(2.5) = 4 + 3(0.5) + 0.5(0.5)^2 + 0(0.5)^3 = 6.625 $$

````python
# пример попроще
import matplotlib.pyplot as plt
from scipy import interpolate

x = np.arange(5, 20)
y = np.exp(x/3.0)

f = interpolate.interp1d(x, y)

x1 = np.arange(6, 12)
y1 = f(x1)   # использовать функцию интерполяции, возвращаемую `interp1d`

plt.plot(x, y, 'o', x1, y1, '--') # o - исходня функция, -- - то, что мы наинтерполировали
plt.show()
````

````python
# пример посложнее
"""
Интерполяция одномерных данных
@author: С.А. Хайбрахманов
"""
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import interp1d

# функция для генерации данных,
# по которым будет строиться интерполяция
def f_exact(x):
 return np.sin(x)**2

# массив координат, по которым будет делаться интерполяция
x = np.linspace(0, 2.0 * np.pi, 10)

# массив значений, по которым будет делаться интерполяция
y = f_exact(x)

# интерполяция данных методом по умолчанию
# (линейная интерполяция)
fi_1 = interp1d(x, y)

# кубическая интерполяция
fi_2 = interp1d(x, y, 'cubic')

# массив координат для построения
# графика интерполированных функций
xi = np.linspace(0, 2.0 * np.pi, 25)

# массивы значений интерполированых функций
# для построения графиков
yi_1 = fi_1(xi)
yi_2 = fi_2(xi)

# настройки осей
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(0.0, 1.5)

# рисование графиков
plt.plot(x, y, 'o', label='исходные данные')
plt.plot(xi, f_exact(xi), '-', color='grey', linewidth=2.5, label='точная функция')
plt.plot(xi, yi_1, '--', label='линейная интерполяция')
plt.plot(xi, yi_2, 'k:', label='кубическая интерполяция')

# легенда, располагающаяся сверху по центру панели
plt.legend(loc='upper center')
plt.show()
````

## Оптимизация (scipy.optimize)

Простыми словами:

- Оптимизация функций в высшей математике — это процесс поиска таких значений переменных, при которых функция достигает своего минимального или максимального значения. Проще говоря, это как игра в "горячо-холодно", где вы пытаетесь найти самое "горячее" (максимум) или "холодное" (минимум) место на карте функции.

- Например, если у вас есть функция, которая описывает стоимость производства в зависимости от количества произведенных товаров, оптимизация поможет вам найти такое количество товаров, при котором стоимость будет наименьшей, то есть вы найдете самый экономичный вариант производства.

- В высшей математике это делается с помощью различных методов и алгоритмов, которые помогают вычислить оптимальные точки, не пробуя каждое возможное значение, что было бы неэффективно, особенно если переменных много¹.

Модуль `scipy.optimize` включает в себя комплекс функций для реализации алгоритмов оптимизации, которые обеспечивают максимизацию или минимизацию функций, аппроксимацию кривых, нахождение корней уравнений и другие задачи оптимизации.

Детальнее о функциональности модуля:

- **Минимизация многомерных скалярных функций**: Включает в себя как неограниченную, так и ограниченную минимизацию. Для этого используются методы, такие как алгоритм Бройдена — Флетчера — Гольдфарба — Шанно (BFGS), метод сопряженных градиентов, метод Нелдера — Мида и другие.
- **Глобальная оптимизация**: Применяются алгоритмы, такие как дифференциальная эволюция и двойной отжиг, для нахождения глобального минимума функции.

В рамках численной оптимизации, функция dual_annealing представляет собой эффективный алгоритм для нахождения глобального минимума целевой функции, особенно когда область определения функции является обширной и метод полного перебора (`brute()`) становится неэффективным из-за высоких вычислительных затрат.

Метод dual_annealing сочетает в себе стратегии классического имитированного отжига и быстрого имитированного отжига, дополненные механизмом локального поиска в принятых решениях. Это позволяет алгоритму эффективно исследовать пространство решений, избегая застревания в локальных минимумах и повышая вероятность достижения глобального оптимума.

Алгоритм dual_annealing работает следующим образом:
1. Инициализация начального состояния и установка начальной температуры*.
2. Генерация нового состояния на основе "визитной" распределительной функции.
3. Оценка нового состояния с помощью "приемной" распределительной функции.
4. Применение локального поиска к принятым новым состояниям для уточнения решения.
5. Постепенное снижение температуры согласно заданному расписанию охлаждения.
6. Повторение процесса до выполнения критериев остановки, таких как достижение максимального числа итераций или минимальной температуры.

*В этих алгоритмах температура представляет собой параметр, который контролирует степень исследования пространства решений. При высокой температуре алгоритм может принимать решения, которые ухудшают целевую функцию, что позволяет ему исследовать более широкий диапазон возможных решений и избегать застревания в локальных минимумах. По мере “охлаждения” системы, или уменьшения температуры, алгоритм начинает сосредотачиваться на улучшении найденных решений, уменьшая вероятность принятия худших решений и сужая поиск к потенциальному глобальному минимуму

````python
xmin_global = optimize.dual_annealing(f, (grid,))
print(xmin_global) # наша точка минимума
````

Далее классический пример минимум функции Розенброка **методом наименьших квадратов**:

- Метод наименьших квадратов (МНК) - это метод оптимизации, который находит параметры функции, наилучшим образом подходящей набору данных. Он делает это путем минимизации суммы квадратов разностей между значениями функции и фактическими значениями данных.

- Функция Розенброка - это нелинейная функция, имеющая несколько локальных минимумов. Она часто используется для тестирования методов оптимизации.

````python
# функция Розенброка
def fun_rosenbrock(x):
   return np.array([10 * (x[1] - x[0] ** 2), (1 - x[0])])

from scipy.optimize import least_squares

input_data = np.array([2, 2])
result = least_squares(fun_rosenbrock, input_data)

print(result)
````

## Функции преобразования Фурье (scipy.fft)

### Простыми словами

**1. Что это:**

Преобразование Фурье – это математический инструмент, который позволяет **разложить** сложную функцию на **набор более простых функций**. Эти простые функции – **синусоиды** с разными частотами.

**2. Зачем это нужно:**

* **Анализ:** Позволяет понять, из каких частот **состоит** сигнал.
* **Фильтрация:** Можно **убрать** ненужные частоты (например, шум).
* **Сжатие:** Можно **уменьшить** размер сигнала, сохранив важные частоты.
* **Обработка:** Можно **упростить** сложные операции с сигналами.

**3. Как это работает:**

* **Разложение:** Преобразование Фурье **раскладывает** сигнал на набор синусоид.
* **Спектр:** Набор синусоид называется **спектром Фурье**.
* **Амплитуда и фаза:** Каждая синусоида имеет **амплитуду** (высоту) и **фазу** (сдвиг).
* **Обратное преобразование:** Можно **сложить** синусоиды обратно, чтобы получить исходный сигнал.

**4. Примеры:**

* **Музыка:** Преобразование Фурье используется для **анализа** и **изменения** музыки.
* **Изображения:** Преобразование Фурье используется для **сжатия** и **обработки** изображений.

**5. Суть:**

- Преобразование Фурье позволяет **перейти** из **временной** области (сигнал во времени) в **частотную** область (спектр Фурье).

SciPy предлагает модуль **fftpack**, который позволяет пользователю быстро вычислять преобразования Фурье. Функции **fft** используются для получения дискретной трансформации Фурье реальной или комплексной последовательности.

**scipy.fftpack** использует:
* scipy.fftpack.fftfreq (): создать последовательность выборки;
* scipy.fftpack.fft (): вычислить быстрое преобразование Фурье.

````python
from scipy.fftpack import fft, ifft

data = np.array([0, 1, 2, 3])

result = fft(data)
print(result)
````

## Линейная алгебра (scipy.linalg)

Модуль scipy.linalg предоставляет широкий спектр инструментов для решения задач линейной алгебры в Python. Часть функций соответствует функциям модуля np.linalg

Функция **scipy.linalg.solve** решает **линейное уравнение** вида $ax+by=Z$ для неизвестных значений $x$, $y$. Рассмотрим на примере. **Решим следующую систему уравнений**: \begin{matrix}
 &  & \\ x+3y+5z=10
 &  & \\ 2x+5y+z=8
 &  & \\ 2x+3y+8z=3
\end{matrix}.

````python
from scipy import linalg
import numpy as np

# записываем коэффиценты левой части
a = np.array([[1, 3, 5],
              [2, 5, 1],
              [2, 3, 8]])

# записываем ответы из правой части
b = np.array([10,
              8,
              3])

result = linalg.solve(a, b)
print(result)
````

Соответственно, x=-9.28, y=5.16, z=0.76.

Теперь попробуем вычислить **определитель матрицы**: $ A=\begin{pmatrix}
 1 & 2 \\
 3 & 4
\end{pmatrix}$

````python
from scipy import linalg
import numpy as np

# матрица
A = np.array([[1,2],
              [3,4]])

deter = linalg.det(A)
print(deter)
````

Функция **scipy.linalg.eig** возвращает **собственные значения и собственные векторы** матрицы. Например, найдём такоевые для $ A=\begin{pmatrix}
 1 & 2 \\
 3 & 4
\end{pmatrix}$

````python
A = np.array([[1,2],
              [3,4]])

l, v = linalg.eig(A)

# собственные значения
print(l)
print()
# векторы
print(v)
````

Функция **scipy.linalg.svd** отвечает за **сингулярное разложение матрицы** на две унитарные матрицы $U$ и $V$ и одномерный массив $s$ сингулярных значений.

````python
from scipy import linalg
import numpy as np

# создадим случайную матрицу
a = np.random.randn(3, 2) + 1.j*np.random.randn(3, 2)
print(a)
print()
# применяем svd
U, s, Vh = linalg.svd(a)

print (U, Vh, s, sep='\n\n')
````

Функция **linalg.inv** позволяет найти **обратную матрицу**:

````python
A = np.array([[1,2],
              [4,3]])
print('Первоначальная матрица A:')
print(A)

B = linalg.inv(A)

print('\nОбратная А матрица:')
print(B)
````

## Разреженные структуры данных и алгоритмы (пространственные) (scipy.spatial)

**1. Что такое разреженные данные:**

Представьте, что у вас есть набор данных о звездах в галактике. Вместо того, чтобы хранить информацию о каждой точке пространства, вы храните только информацию о звездах. Это означает, что большинство данных "пустые", а информация "разрежена".

**2. scipy.spatial:**

Библиотека scipy.spatial содержит инструменты для работы с разреженными данными. Она позволяет:

* **Вычислять триангуляции:** Делает из набора точек "сетку" из треугольников.
* **Строить диаграммы Вороного:** Разбивает пространство на ячейки,
    * Каждая ячейка принадлежит
    * Ближайшей к ней точке.
* **Находить выпуклые оболочки:** Находит "минимальную" оболочку,
    * Которая охватывает
    * Все точки.
* **Искать ближайших соседей:** Быстро находит
    * Ближайшие к данной точке
    * Другие точки.
* **Вычислять расстояния:** Измеряет
    * Расстояние между
    * Любыми двумя точками.

**3. Примеры использования:**

* **Астрономия:** Анализ распределения звезд в галактике.
* **География:** Поиск оптимальных маршрутов на карте.
* **Компьютерное зрение:** Распознавание объектов на изображениях.
* **Машинное обучение:** Кластеризация данных.

**4. KDTree:**

KDTree - это структура данных,
* Которая используется для быстрого поиска ближайших соседей.

**5. Метрики:**

Метрики - это способы измерения расстояния между точками.
* Евклидово расстояние -
* Самый распространенный
* Пример метрики.

- Рассмотрим пример **триангуляции Делоне**. Триангуляции Делоне - это точки на плоской поверхности — это триангуляция, где ни одна точка из представленных не расположена внутри описанной окружности любого треугольника.

````python
import matplotlib.pyplot as plt
from scipy.spatial import Delaunay

# точки
points = np.array([[0, 1], [1, 1], [1, 0],[0, 0]])

a = Delaunay(points)       # Объект Делоне
print(a)
print(a.simplices)

plt.triplot(points[:,0], points[:,1], a.simplices)
plt.plot(points[:,1], points[:,0], 'o')
plt.show()
````

## Алгоритмы кластерного анализа (scipy.cluster)

**scipy.cluster** содержит набор алгоритмов кластеризации, например, **K-means**. Кластеризация K-средних – это метод поиска кластеров и центров кластеров в наборе немаркированных данных. Интуитивно, мы можем думать о кластере как о – состоящем из группы точек данных, чьи расстояния между точками малы по сравнению с расстояниями до точек вне кластера.

````python
from scipy.cluster.vq import kmeans, vq, whiten
from numpy import vstack,array
from numpy.random import rand

# сгенерируем сдучайные данные
data = vstack((rand(100,3) + array([.5,.5,.5]),rand(100,3)))
print(data[:5], '\n')

# почистим данные
data = whiten(data)

# вычисляем K-Means с 3 кластерами
centroids, _ = kmeans(data, 3)
print(centroids, '\n')

# разделим значения по пренадлежности к кластеру
clx, _ = vq(data, centroids)
print(clx)
````

## Compressed Sparse Graph (scipy.sparse.csgraph)

**scipy.sparse.csgraph** включает алгоритмы для работы с **графами**. Например, поиск кратчайшего пути в графе через алгоритм **Дейкстры**.

````python
import numpy as np
from scipy.sparse.csgraph import dijkstra
from scipy.sparse import csr_matrix

# граф
arr = np.array([
  [0, 1, 2],
  [1, 0, 0],
  [2, 0, 0]
])

newarr = csr_matrix(arr)

print(dijkstra(newarr, return_predecessors=True, indices=0))
````

Поддерживается множество алгоритмов в их числе **Алгоритм Флойда — Уоршелла, Беллмана — Форда, поиск в глубину, ширину** и т.д. Давайте на примере и посмотрим **поиск в глубину**:

````python
from scipy.sparse import csr_matrix
from scipy.sparse.csgraph import depth_first_order

graph = [
[0, 1, 2, 0],
[0, 0, 0, 1],
[2, 0, 0, 3],
[0, 0, 0, 0]
]

graph = csr_matrix(graph)
print(graph)

print(depth_first_order(graph, 0))
````
